```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(-10, 11)
plt.figure(figsize=(12,6))
plt.title('title')

plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.plot([0,0,0],[-10,0,100])
plt.legend(['X^2','Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1*(x**2))
plt.plot([-10,0,10],[-50,-50,-50])
plt.legend(['-X^2','Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
plt.show()
```

## Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2
```

```python
plt.figure(figsize=(14,6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

## Histograms

```python
values = np.random.randn(1000)
```

```python
plt.subplots(figsize=(12,6))
plt.hist(values, bins=100, alpha=0.8,
        histtype='bar', color='steelblue',
        edgecolor='green')
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

## KDE(kernel density estimation)

```python
from scipy import stats
density = stats.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12,6))
values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='green')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='green')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

## Combine plots

```python
plt.subplots(figsize=(12,6))
plt.hist(values, bins=100, alpha=0.8, density=1,
        histtype='bar', color='steelblue',
        edgecolor='green')
plt.plot(values2, density(values2), color='orange', linewidth=3)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

## Bar plots

```python
Y = np.random.rand(1,5)[0]
Y2 = np.random.rand(1,5)[0]
```

```python
plt.figure(figsize=(12,4))
barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='blue')
plt.show()
```

```python
plt.figure(figsize=(12,4))
barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='blue', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='green', label='Label Y', bottom=Y)

plt.legend()
plt.show()
```
